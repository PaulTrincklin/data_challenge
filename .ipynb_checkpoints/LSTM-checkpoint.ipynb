```python
def get_data(X_data,y_data):
    y=y_data.to_numpy()
    cols = X_data.columns
    categories = sorted(list(set(list(map(lambda x:x.split('-')[0],cols)))))
    relevant_cat = categories[18:-14]+categories[:12]
    relevant_cat_48 = [x for l in list(map(lambda x:[f"{x}-{i+1}" for i in range(48)],relevant_cat)) for x in l]
    X=np.nan_to_num(X_data[relevant_cat_48].select_dtypes(include='number').to_numpy())
    return X,y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize,MinMaxScaler,StandardScaler

model = tf.keras.models.Sequential([
    # Shape [batch, time, features] => [batch, time, lstm_units]
    tf.keras.layers.LSTM(30, return_sequences=False,input_shape=(48, 23),dropout=0.2),
    # Shape => [batch, time, features]
    tf.keras.layers.Dense(12)
])

optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
model.compile(loss='mse', optimizer=optimizer,metrics=['mse'])

X_train,y_train = get_data(X_data,y_data)
#scaler2 = MinMaxScaler()
#y = scaler2.fit_transform(y)

#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
#X_test = scaler.transform(X_test)

Xr_train = X_train.reshape(X_train.shape[0],48,X_train.shape[1]//48)
#Xr_test = X_test.reshape(X_test.shape[0],48,X_test.shape[1]//48)
```

```python
model.fit(Xr_train,y_train,epochs=500,batch_size=512)
```

```python
from sklearn.metrics import mean_squared_error

X_test,y_true = get_data(X_sub,y_data)
X_test = scaler.transform(X_test)
Xr_test = X_test.reshape(X_test.shape[0],48,X_test.shape[1]//48)
y_pred = model.predict(Xr_test)
#mean_squared_error(y_pred,y_test)
#mean_squared_error(scaler2.inverse_transform(y_pred),scaler2.inverse_transform(y_test))
```
