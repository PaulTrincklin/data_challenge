```python
Xt=np.nan_to_num(X_data[so2].select_dtypes(include='number').to_numpy())
Xt=Xt.reshape(6089*6,48)
Xt
```

```python
yt=np.hstack((y,y,y,y,y,y))
yt=yt.reshape(6089*6,12)
yt
```

```python
from darts import TimeSeries

data = []
for i in range(len(Xt)):
    data.append(TimeSeries.from_values(np.hstack((Xt[i],yt[i]))))
data[0].plot()
data[300].plot()
data[600].plot()
```

```python
#https://unit8co.github.io/darts/examples/04-RNN-examples.html
from darts.models import NBEATSModel

l = int(4*len(Xt)//5)
train = data[:l]
val = data[l:]
model = NBEATSModel(input_chunk_length=48, output_chunk_length=12,n_epochs=5)
model.fit(train)
```

```python
X_val = list(map(TimeSeries.from_values,np.hstack((Xt,yt))[l:][:,:48]))
y_val = list(map(TimeSeries.from_values,np.hstack((Xt,yt))[l:][:,48:]))
total = list(map(TimeSeries.from_values,np.hstack((Xt,yt))[l:]))
forecast = model.predict(12,X_val)
```
